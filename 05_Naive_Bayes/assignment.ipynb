# **Performing Naive Bayes on tips dataset**

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.columns
```

```python
X = df[['total_bill', 'tip', 'sex', 'day', 'time', 'size']]
```

```python
y=df[['smoker']]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size=0.25, random_state=10)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
y_train
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
X_train['sex']=le1.fit_transform(X_train['sex'])
X_train['day'] = le2.fit_transform(X_train['day'])
X_train['time']=le3.fit_transform(X_train['time'])

X_test['sex']=le1.transform(X_test['sex'])
X_test['day'] = le2.transform(X_test['day'])
X_test['time']=le3.transform(X_test['time'])
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
```

```python
acs=  accuracy_score(y_test, y_pred)
clr=classification_report(y_test,y_pred)
cm=confusion_matrix(y_test,y_pred)
print("accurancy score: \n",acs)
print('confusion matrix\n',cm)
print('classification report:\n',clr)
```

Testing on New Data

```python
new_data = [[20.5, 3.5, 1, 2, 0, 2]]
prediction = gnb.predict(new_data)
```

```python
print(prediction)
```

Convert the prediction back to the original label

```python
result = le.inverse_transform(prediction)
print(result)
```
